```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# Web scraping from wikipedia
def load_data():
    url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
    html = pd.read_html(url, header = 0)
    df = html[0]
    return df

df=load_data()
```

```python
data = yf.download(
    tickers = list(df.Symbol),
    period = "ytd",
    interval = "1d",
    group_by = 'ticker'
)
```

```python
data
```

```python
data['AES']
```

```python
df_Close = pd.DataFrame(data['AES'].Close)
df_Close['Date']= df_Close.index
df_Close
```

```python
def price_plot(sym):
    df_Close = pd.DataFrame(data[sym].Close)
    df_Close['Date']= df_Close.index
    plt.fill_between(df_Close.Date, df_Close.Close, color='gold', alpha=0.2)
    plt.plot(df_Close.Date, df_Close.Close, color='gold', alpha=0.8)
    plt.xticks(rotation=90)
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.title(f'{sym}', fontweight='bold')
    return plt.show()
```

```python
price_plot('ISRG')
```

```python
for sym in list(df.Symbol)[:5]:
    price_plot(sym)
```
